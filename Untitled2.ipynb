```python
#displot -- i have no idea what these are 
#joinplot
#pairplot
#rugpull
#kdeplot
import seaborn as seamen # seamen adds authenticity to our program 
%matplotlib inline
```

```python
tips  = seamen.load_dataset('tips')
```

```python
print(tips)
```

```python
tips.head()
```

```python
seamen.displot(tips['total_bill'])
```

```python
# just have the histogram 
seamen.displot(tips['total_bill'],kde= False,bins=10 ) # kde means kernel destiny estimate 
```

```python
seamen.jointplot(x = 'total_bill', y='tip', data=tips,kind='scatter' )
```

```python
# regression plots 

anotherTip = seamen.load_dataset('tips') 
seamen.lmplot('total_bill',y='tip', data= anotherTip)
```

```python
seamen.lmplot('total_bill',y='tip', data=anotherTip, hue='sex')
```

```python
seamen.lmplot('total_bill',y='tip', data= anotherTip, col='sex')
```
